```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'State Political Standing':['Strong Rep', 'Strong Dem', 'Weak Rep', 'Weak Dem'], 'val':[43, 37, 44, 45]})

ax = df.plot.bar(x='State Political Standing', y='val', color=['red', 'blue', 'red', 'blue'], rot=0, legend=None)

plt.title('Under-Reporting by Party Affiliation')
plt.xlabel('State Political Standing')
plt.ylabel('Avg. Percentage of Under-Reported Covid Deaths')
plt.tight_layout()
plt.show()
```
