```python
import pandas as pd
import os
import re
import numpy as np
```

```python
df = pd.read_excel(f"Resources/aacer-sep-2020-commercial-bankruptcy-filings-all-chapters-ch-11-focus.xlsx", 
                   sheet_name="Chapter 11 Summary")
```

```python
month_header = pd.Series(df.iloc[0,0])
month_header = list(month_header)

year_headers = pd.Series(df.iloc[1,1:])
year_headers = list(year_headers.values)

df_labels = month_header + year_headers

df.columns = df_labels

df = df.iloc[2:,:]

df = df.reset_index(drop=True)

df = df.drop([0, 13, 14, 15])

df
```

```python
df = pd.melt(df, id_vars=["Month"], var_name="Year", value_name = "Bankruptcies")

df['Year'] = df['Year'].astype(float)
df['Year'] = df['Year'].astype(int) 

df['MonthInitial'] = df['Month'].str[:3]

df['Month-Year'] = df.MonthInitial + "-" + df.Year.map(str)

df.head()
```

```python
df['Month-Year'] = pd.to_datetime(df['Month-Year'], format='%b-%Y')
df
```

```python
df = df.drop(['Month', 'Year', 'MonthInitial'], axis = 1)

df = df.sort_values(by=['Month-Year'])

df = df.set_index("Month-Year")

print(f"Monthly Chapter 11 Bankruptcies Jan 2011 thru Sep 2020")
df.head()
```
