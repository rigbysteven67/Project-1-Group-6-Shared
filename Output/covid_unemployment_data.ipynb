```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cases_file = '../Resources/Covid cases by state.csv'
cases_df = pd.read_csv(cases_file, encoding="ISO-8859-1")
cases_df = cases_df.fillna(0)

revised_cases_df = cases_df[['state','date','total_population','new_cases','new_cases_per_100_000','cumulative_cases','cumulative_cases_per_100_000']]
revised_cases_df['% of New Cases'] = (revised_cases_df['new_cases'] / revised_cases_df['total_population']) *100
revised_cases_df = revised_cases_df.fillna(0)
revised_cases_df.head()
```

```python
unemployment_file = '../Resources/Unemployment claims by state.csv'
unemployment_df = pd.read_csv(unemployment_file, encoding="ISO-8859-1")
unemployment_df = unemployment_df.fillna(0)

revised_unemployment_df = unemployment_df.rename(columns={'Filed week ended':'date', 'ï»¿State': 'state'})
revised_unemployment_df = revised_unemployment_df[['state', 'date', 'Initial Claims', 'Insured Unemployment Rate']]
revised_unemployment_df = revised_unemployment_df.fillna(0)
revised_unemployment_df.head()
```

```python
claims_with_cases_df = pd.merge(revised_cases_df, revised_unemployment_df, on=['date','state'], how='inner')
claims_with_cases_df = claims_with_cases_df.fillna(0)
claims_with_cases_df = claims_with_cases_df[['date','state','total_population','new_cases','new_cases_per_100_000','cumulative_cases','cumulative_cases_per_100_000','Initial Claims','Insured Unemployment Rate']]
claims_with_cases_df.head()
```

```python
# Path of csv file, originally converted using tabula
state_political_control_path = "../Resources/Legislative Partisan Composition by State 2020-04-01 - Processed.csv"

# Read the csv file
state_political_control_df = pd.read_csv(state_political_control_path, encoding="ISO-8859-1")
```

```python
# Display the data table for preview
# Replace all NaN values with 0 
state_political_control_df = state_political_control_df.fillna(0)

state_political_control_df = state_political_control_df.rename(columns={'ï»¿STATE': 'state'})

state_political_control_df = state_political_control_df[['state', 'State Control']]

state_political_control_df.loc[state_political_control_df['State Control'] == 0, 'State Control'] = 'Non_Partisan'
state_political_control_df.head()
```

```python
summary_political_control_df = state_political_control_df['State Control'].value_counts()

fig = plt.figure(figsize=(10,5))
ax =  state_political_control_df['State Control'].value_counts().plot(kind='bar',
                                                                        figsize=(10,5),
                                                                        color=['red', 'blue', 'yellow', 'gray'],
                                                                        title="Count of Partisan Group Control by State")
plt.xticks(rotation=45)
ax.set_xlabel("Partisan Groups")
ax.set_ylabel("# Number of States")

yint = []
locs, labels = plt.yticks()
for each in locs:
    yint.append(int(each))
plt.yticks(yint)

fig.savefig('Count of Partisan Group Control by State.jpg', bbox_inches='tight', dpi=150)
```

```python
combined_df = pd.merge(claims_with_cases_df, state_political_control_df, on=['state'], how='inner')
combined_df['Initial Claims'] = combined_df['Initial Claims'].str.replace(',', '').astype(float)
combined_df['date'] = combined_df['date'].astype('datetime64[ns]')
combined_df.loc[combined_df['State Control'] == 0, 'State Control'] = 'Non_Partisan'
combined_df = combined_df.fillna(0)
combined_df.head()
```

```python
party_grouped_df = combined_df.groupby(['State Control', 'date'])
party_summary_df = pd.DataFrame({
    'Total # of New Cases per 100,000':party_grouped_df['new_cases_per_100_000'].sum(),
    'Avg # of New Cases per 100,000':party_grouped_df['new_cases_per_100_000'].mean(),
    'Avg. Rate':party_grouped_df['Insured Unemployment Rate'].mean()
})
party_summary_df
```

```python
party_summary_df = party_summary_df.unstack(level=0)

dates = party_summary_df.index.values

new_cases_dem = party_summary_df['Total # of New Cases per 100,000']['Dem']
new_cases_rep = party_summary_df['Total # of New Cases per 100,000']['Rep']
new_cases_div = party_summary_df['Total # of New Cases per 100,000']['Divided']
new_cases_nonp = party_summary_df['Total # of New Cases per 100,000']['Non_Partisan']

fig = plt.figure(figsize=(10,5))
dem_cases, = plt.plot(dates, new_cases_dem, color="blue", label="Democratic Control",)
rep_cases, = plt.plot(dates, new_cases_rep, color="red", label="Republican Control")
div_cases, = plt.plot(dates, new_cases_div, color="yellow", label="Divided Control")
nonp_cases, = plt.plot(dates, new_cases_nonp, color="gray", label="Non-Partisan Control")
plt.legend()
plt.title('Total Number of New Covid Cases by State Level Partisan Control per 100,000')
plt.xlabel('Date')
plt.ylabel('Total # of New Cases (per 100,000)')
plt.xticks(rotation=45)
fig.savefig('Total Number of New Covid Cases by State Level Partisan Control per 100,000.jpg', bbox_inches='tight', dpi=150)
```

```python
avg_new_cases_dem = party_summary_df['Avg # of New Cases per 100,000']['Dem']
avg_new_cases_rep = party_summary_df['Avg # of New Cases per 100,000']['Rep']
avg_new_cases_div = party_summary_df['Avg # of New Cases per 100,000']['Divided']
avg_new_cases_nonp = party_summary_df['Avg # of New Cases per 100,000']['Non_Partisan']

fig = plt.figure(figsize=(10,5))
avg_dem_cases, = plt.plot(dates, avg_new_cases_dem, color="blue", label="Democratic Control",)
avg_rep_cases, = plt.plot(dates, avg_new_cases_rep, color="red", label="Republican Control")
avg_div_cases, = plt.plot(dates, avg_new_cases_div, color="yellow", label="Divided Control")
avg_nonp_cases, = plt.plot(dates, avg_new_cases_nonp, color="gray", label="Non-Partisan Control")
plt.legend()
plt.title('Average Number of New Covid Cases by State Level Partisan Control per 100,000')
plt.xlabel('Date')
plt.ylabel('Avg # of New Cases (per 100,000)')
plt.xticks(rotation=45)
fig.savefig('Average Number of New Covid Cases by State Level Partisan Control per 100,000.jpg', bbox_inches='tight', dpi=150)
```

```python
unemployment_dem = party_summary_df['Avg. Rate']['Dem']
unemployment_rep = party_summary_df['Avg. Rate']['Rep']
unemployment_div = party_summary_df['Avg. Rate']['Divided']
unemployment_nonp = party_summary_df['Avg. Rate']['Non_Partisan']

fig = plt.figure(figsize=(10,5))
dem_unemployment, = plt.plot(dates, unemployment_dem, color="blue", label="Democratic Control",)
rep_unemployment, = plt.plot(dates, unemployment_rep, color="red", label="Republican Control")
div_unemployment, = plt.plot(dates, unemployment_div, color="yellow", label="Divided Control")
nonp_unemployment, = plt.plot(dates, unemployment_nonp, color="gray", label="Non-Partisan Control")
plt.legend(loc='upper right')
plt.title('Unemployment Rate by State Level Partisan Control')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate %')
plt.xticks(rotation=45)
fig.savefig('Unemployment Rate by State Level Partisan Control.jpg', bbox_inches='tight', dpi=150)
```

```python
state_grouped_df = combined_df.groupby(['state', 'date'])
state_summary_df = pd.DataFrame({
    'Total # of New Cases per 100,000':state_grouped_df['new_cases_per_100_000'].sum(),
    'Avg # of New Cases per 100,000':state_grouped_df['new_cases_per_100_000'].mean(),
    'Avg. Rate':state_grouped_df['Insured Unemployment Rate'].mean()
})
state_summary_df

```

```python
state_summary_df = state_summary_df.unstack(level=0)

state = (input('Enter a state: ').title())

new_cases_state = state_summary_df['Total # of New Cases per 100,000'][state]

fig = plt.figure(figsize=(10,5))
state_cases, = plt.plot(dates, new_cases_state, color="orange", label=state,)

plt.legend()
plt.title(f'Total Number of New Covid Cases per 100,000: {state}')
plt.xlabel('Date')
plt.ylabel('Total # of New Cases (per 100,000)')
plt.xticks(rotation=45)
fig.savefig(f'Total Number of New Covid Cases per 100,000: {state}.jpg', bbox_inches='tight', dpi=150)
```

```python
unemployment_state = state_summary_df['Avg. Rate'][state]

fig = plt.figure(figsize=(10,5))
state_unemployment, = plt.plot(dates, unemployment_state, color="orange", label=state,)

plt.legend()
plt.title(f'Unemployment Rate: {state}')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate %')
plt.xticks(rotation=45)
fig.savefig(f'Unemployment Rate: {state}.jpg', bbox_inches='tight', dpi=150)
```
